# Load necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Iris dataset and display first few rows

```python
iris = sns.load_dataset('iris')
iris.head()
```

# Check for missing values and show basic statistics

```python
iris.info()
```

```python
# Null values
iris.isnull().sum()
```

```python
# Basic statistics for dataset
iris.describe()
```

## Univariate Analysis: Distribution of one variable



```python
plt.figure(figsize=(10, 5))
sns.histplot(iris['sepal_length'], kde=True, color='blue')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

## Univariate Analysis: Boxplot
**Useful to check for outliers in the data**

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Boxplot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

## Bivariate Analysis: Distribution of Two Variables

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Scatterplot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.show()
```

```python
# Pairplot

sns.pairplot(iris, hue='species', diag_kind='kde')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
# Correlation Analysis
correlation_matrix = iris.drop(columns=['species']).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```

```python
# Additional: Violin plot for detailed distribution by species
plt.figure(figsize=(10, 5))
sns.violinplot(x='species', y='petal_length', data=iris)
plt.title('Violin Plot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()
```

# Conclusions Section

1. Sepal and petal dimensions differ significantly between species.
2. Sepal width and length show weak correlation.
3. Petal dimensions are highly correlated.
